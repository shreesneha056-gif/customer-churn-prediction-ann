```python
from scikeras.wrappers import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense
import optuna
from optuna_integration import OptunaSearchCV
```

```python
def create_model(n_layers=1):
    model = Sequential()
    model.add(Dense(units=6, activation='relu', input_dim=9)) 
    
    for i in range(n_layers - 1):
        model.add(Dense(units=6, activation='relu'))
        
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
KerasClassifier?
```

```python
wrapped_classifier = KerasClassifier(model=create_model, n_layers=1, verbose=0)
```

```python
param_distributions = {
    "model__n_layers": optuna.distributions.IntDistribution(1, 3),
    "optimizer__learning_rate": optuna.distributions.FloatDistribution(1e-5, 1e-1, log=True)
}
```

```python
OptunaSearchCV?
```

```python
optuna_search = OptunaSearchCV(
    estimator=wrapped_classifier, 
    param_distributions=param_distributions, 
    n_trials=5, 
    cv=3,
    error_score='raise'
)
```

```python
optuna_search.fit(X_train, y_train)
```

```python
optuna_search.best_params_
```

```python
optuna_search.best_score_
```

```python
best_model = optuna_search.best_estimator_
```

```python
y_pred_test_cv = best_model.predict(X_test)
y_pred_test_cv_optimized = (y_pred_test_cv > 0.5)
```

```python
y_pred_train_cv = best_model.predict(X_train)
y_pred_train_cv_optimized = (y_pred_train_cv > 0.5)
```

```python
print(classification_report(y_train, y_pred_train_cv_optimized))
print(classification_report(y_test, y_pred_test_cv_optimized))
```
